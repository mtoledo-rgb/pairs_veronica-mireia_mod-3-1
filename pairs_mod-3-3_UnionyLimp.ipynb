```python
nuevas_columnas1 = {columna: columna.lower() for columna in df_1.columns}
nuevas_columnas2 = {columna: columna.lower() for columna in df_2.columns}
```

```python
df_1.rename(columns = nuevas_columnas1, inplace = True)
df_2.rename(columns = nuevas_columnas2, inplace = True)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_final=df_1.merge(df_2, how="inner", on="country")
```

```python
df_final.head()
```

Identificamos que la unión se podia realizar a través de un Merge, por que tenian en común la columna "country", sin embargo, una estaba con mayúscula inicial "Country" y otra todo minuscula, por lo que debemos aplicarle un rename a las columnas. 

```python
# No tengan espacios.
sin_espacios = {columna: columna.replace(" ", "") for columna in df_final.columns}
df_final.rename(columns = sin_espacios, inplace = True)
```

```python
df_final.head()
```

```python
# No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
sin_parentesis1 = {columna: columna.replace("(km2)", "") for columna in df_final.columns}
sin_parentesis2 = {columna: columna.replace("(%)", "") for columna in df_final.columns}
sin_parentesis3= {columna: columna.replace("(p/km2)", "") for columna in df_final.columns}
df_final.rename(columns = sin_parentesis1, inplace = True)
df_final.rename(columns = sin_parentesis2, inplace = True)
df_final.rename(columns = sin_parentesis3, inplace = True)
```

```python
df_final.head()
```

```python
# Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
sin_n= {columna: columna.replace("density\n", "density") for columna in df_final.columns}
df_final.rename(columns = sin_n, inplace = True)
```

```python
df_final.head()
```

```python
# Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.
sin_puntos= {columna: columna.replace("population:laborforceparticipation", "populationlaborforceparticipation") for columna in df_final.columns}
df_final.rename(columns = sin_puntos, inplace = True)
```

```python
df_final.head()
```

```python
# codigo que nos pasó césar, para simplificar esto 
columnas = df_final.columns
nuevas_columnas = {col: col.split("(")[0].strip().replace(" ", "_").replace("\n", "").replace(":", "").lower() for col in columnas}
df_final.rename(columns=nuevas_columnas, inplace = True)
```
