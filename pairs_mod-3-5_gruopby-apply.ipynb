```python
def num(dato):
    try:
        return float(dato.replace(',','.').replace('$',''))
    except:
        return np.nan
```

```python
lista1 = ['land_area', 'armed_forces_size', 'co2-emissions', 'gdp', 'population', 'urban_population', 'latitude', 'longitude', 'density']
```

```python
for col in lista1:
    df_full[col] = df_full[col].apply(num).astype(float)  #astype para pasarlo al tipo de dato en el que queremos convertirlo
```

```python
#2. Crea una función para añadir una nueva columna llamada 'Continent' 
#al DataFrame para indicar a qué continente pertenece cada país.
```

```python
def continent(datos):
    if datos in dicc_continent: #si el dato aparece en el diccionario, siendo "dato"='country'
        return dicc_continent[datos] #me devuelve el continente al que pertenece
```

```python
df_full['continent']=df_full['country'].apply(continent) #aquí se aplica la función enlazándola con la columna que ya existe de 'country'
df_full[['continent','country']].head(3)
```

```python
#1. ¿Cuál es la densidad de población promedio por continente?
```

```python
df_final.groupby('continent')['density'].mean(numeric_only=True).sort_values().reset_index()
```

```python
#2. ¿Cuál es la esperanza de vida promedio por idioma oficial? 
# Agrupa los países por idioma oficial y calcula la esperanza de vida promedio.
```

```python
df_final.groupby(['country','official_language'])['life_expectancy'].mean(numeric_only=True).sort_values().reset_index()
```

```python
#3. ¿Cuál es la tasa de mortalidad infantil promedio por moneda?
```

```python
df_final.groupby('currency-code')['infant_mortality'].mean(numeric_only=True).sort_values(ascending=False).reset_index()
```
